```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('stock_prices3.csv')
```

```python
data.head()
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data.set_index('date', inplace=True)
```

```python
data['30_day_moving_avg'] = data['closing_price'].rolling(window=30).mean()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['closing_price'], label='Closing Price')
plt.plot(data.index, data['30_day_moving_avg'], label='30-Day Moving Avg', linestyle='dashed')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Prices and 30-Day Moving Average')
plt.legend()
plt.grid(True)
plt.show()
```
